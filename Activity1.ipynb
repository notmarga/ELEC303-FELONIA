<h1>Filling the null values</h1>
<hr>

```python
penguins['bill_depth_mm'] = penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean())
penguins['bill_length_mm'] = penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean())
penguins['flipper_length_mm'] = penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].mean())
penguins['body_mass_g'] = penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean())
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
```

```python
penguins.isnull().sum()
```

```python
penguins.info()
```

<h1>Changing the Data type</h1>
<hr>

```python
penguins['species'] = penguins['species'].astype('category')
penguins['island'] = penguins['island'].astype('category')
penguins['sex'] = penguins['sex'].astype('category')
```

```python
penguins.info()
```

<h1>Creating your own column</h1>
<hr>

```python
penguins['mass_kg'] = np.around(penguins['body_mass_g'] / 1000, 1)
```

```python
penguins
```

<h2>Question 1</h2>
<hr>
<h3>How many Chinstrap have a flipper lenght greater than 210g?</h3>

```python
pgns_df = penguins[(penguins['species'] == 'Chinstrap') & (penguins['flipper_length_mm'] >= 210)]
```

```python
pgns_df.value_counts().sum()
```

```python
pgns_df 
```

<h5>INSIGHT:   All 4 Chinstrap penguins with flipper length is greater than 210 mm are male and located on Dream Island. This suggests that within the Chinstrap species, males may develop longer flippers.</h5>

<h2>Question 2</h2>
<h3>How many penguins from Torgersen Island have a body mass of less than 3200 g?</h3>

```python
pgns = penguins[(penguins['island'] == 'Torgersen') & (penguins['body_mass_g'] <= 3200)]
```

```python
pgns.value_counts().sum()
```

```python
pgns
```

<h5> INSIGHT: All 7 penguins with body mass less than 3200 g from Torgersen Island are female Adelie penguins. This indicates that Adelie females on Torgersen are typically smaller, and Torgersen may host one of the lightest penguin populations in the dataset.
</h5>
